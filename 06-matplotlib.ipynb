```python
import matplotlib.pyplot as plt
plt.rcParams["font.sans-serif"]=["SimHei"]#设置字体
plt.rcParams["axes.unicode_minus"]=False#该语句解决图像中的"-"负号的乱码问题
```

```python
x = [1,2,3,4]
y = [1,4,2,3]
```

```python
fig, ax = plt.subplots()
ax.plot(x,y)
```

```python
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
plt.show()
```
